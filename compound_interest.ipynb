## Launch Web Dashboard

Click "Run" below to start the web server and open the interactive dashboard in your browser.


```python
import webbrowser
import threading
import time
import http.server
import socketserver
import os

PORT = 8000
os.chdir('h:\\Doks\\Projects\\cld')

class MyHTTPRequestHandler(http.server.SimpleHTTPRequestHandler):
    def log_message(self, format, *args):
        return

def start_server():
    try:
        with socketserver.TCPServer(("", PORT), MyHTTPRequestHandler) as httpd:
            httpd.serve_forever()
    except OSError:
        pass

server_thread = threading.Thread(target=start_server, daemon=True)
server_thread.start()

time.sleep(1)

try:
    webbrowser.open(f'http://localhost:{PORT}/')
    print(f"✅ Dashboard opened: http://localhost:{PORT}/")
except:
    pass

print(f"✅ Web server running at: http://localhost:{PORT}/")
```
